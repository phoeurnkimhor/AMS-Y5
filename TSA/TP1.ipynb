1. Use software to produce the time series plot shown in Exhibit 1.2, on page 2. The
data are in the file named ```larain```.

````r
# install.packages("TSA")
library(TSA)
data(larain)
par(bg = "white")
plot(y=larain,x=zlag(larain),ylab='Inches', xlab='Previous Year Inches')
````

3. Simulate a completely random process of length 48 with independent, normal values.
Plot the time series plot. Does it look “random”? Repeat this exercise several times
with a new simulation each time.

````r
par(bg = "white")
plot(ts(rnorm(n=48)), type='o', set.seed(42))
````

4. Simulate a completely random process of length 48 with independent, chi-square
distributed values, each with 2 degrees of freedom. Display the time series plot.
Does it look “random” and nonnormal? Repeat this exercise several times with a
new simulation each time.

````r
par(bg = "white")
plot(ts(rchisq(n=48, df=2)), type='o', set.seed(42))
````

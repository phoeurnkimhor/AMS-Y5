1. Create class Library for a library to keep track of their book. For
each book, we want to know many copy does the library still has on shelf. If a
copy is borrowed, who borrow it, what is their contact, phone and email, when
did it was borrow and when it is expected to be return.

```python
class Library:
    def __init__(self):
        self.books = {}

    def add_book(self, title, author, copies):
        self.books[title] = {
            "author": author,
            "total": copies,
            "available": copies,
            "borrowers": []
        }
        print(f"Added '{title}' by {author} ({copies} copies).")

    def show_books(self):
        print("\nLibrary Collection:")
        for title, info in self.books.items():
            print(f"{title} by {info['author']} — Available: {info['available']}/{info['total']}")

    def borrow_book(self, title, name, phone, email):
        if title not in self.books:
            print(f"'{title}' not found in library.")
            return

        book = self.books[title]
        if book["available"] > 0:
            book["available"] -= 1
            book["borrowers"].append({"name": name, "phone": phone, "email": email})
            print(f"{name} borrowed '{title}'.")
        else:
            print(f"No copies of '{title}' are available.")

    def return_book(self, title, name):
        if title not in self.books:
            print(f"'{title}' not found in library.")
            return

        book = self.books[title]
        for record in book["borrowers"]:
            if record["name"] == name:
                book["borrowers"].remove(record)
                book["available"] += 1
                print(f"{name} returned '{title}'.")
                return
        print(f"No record found for {name} borrowing '{title}'.")

library = Library()

library.add_book("Python Crash Course", "Eric Matthes", 3)
library.add_book("Clean Code", "Robert C. Martin", 2)

library.show_books()

library.borrow_book("Python Crash Course", "Alice", "012345678", "alice@email.com")
library.borrow_book("Python Crash Course", "Bob", "098765432", "bob@email.com")

library.show_books()

library.return_book("Python Crash Course", "Alice")
library.show_books()
```

```python
import torch

A = torch.tensor([[1, 2, 3, 4], [4, 3, 2, 1], [5, 6, 7, 8]])
B = torch.tensor([[1, 0], [0, 1], [1, 0], [0, 1]])

C = A @ B

print("="*50)
print(f"Tensor A: {A}")
print(f"Shape of Tensor A: {A.shape}")
print("="*50)
print(f"Tensor B: {B}")
print(f"Shape of Tensor B: {B.shape}")
print("="*50)
print(f"Tensor C (A @ B): {C}")
print(f"Shape of Tensor C (A @ B): {C.shape}")
print("="*50)
```

5. Create a tensor x representing 200 points from 0 to 2π. Compute y =

```python
x = torch.linspace(0, 2*torch.pi, 200)
y = torch.sin(x)

import matplotlib.pyplot as plt
plt.plot(x, y)
plt.show()
```

7. On a single graph, plot the functions y1 = cos(x) and y2 = cos(2x) for
x ranging from 0 to 4π. Add a legend to your plot to distinguish be-
tween the two curves. (Hint: you can call plt.plot() twice before calling
plt.show()).

```python
x = torch.linspace(0, 4*torch.pi, 200)
y1 = torch.cos(x)
y2 = torch.cos(2*x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.legend(["Cos(x)", "Cos(2x)"])
plt.show()
```
